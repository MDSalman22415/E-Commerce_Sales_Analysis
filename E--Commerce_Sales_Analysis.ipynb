# Converting data columns

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Order Date'].dtype
```

```python
data.info()
```

```python
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
data['Ship Date'].dtype
```

```python
data.info()
```

```python
data.head()
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order Day of Week'] = data['Order Date'].dt.dayofweek
```

```python
data.head()
```

# Monthly-sales analysis

```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
sales_by_month
```

```python
fig = px.line(sales_by_month,
             x='Order Month',
             y='Sales',
             title= 'Monthly Sales Analysis')
fig.show()
```

```python
data.head()
```

# Sales by Category

```python
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
sales_by_category
```

```python
fig =px.pie(sales_by_category,
           values='Sales',
           names='Category',
           hole=0.5,
           color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Sales Analysis by Category', title_font=dict(size=24))

fig.show()
```

# Sales analysis by Sub Category

```python
data.head()
```

```python
sales_by_subcategory = data.groupby('Sub-Category')['Sales'].sum().reset_index()
sales_by_subcategory
```

```python
fig = px.bar(sales_by_subcategory,
            x='Sub-Category',
            y='Sales',
            title='Sales analysis by sub category')
fig.show()
```

# sales and profit customer segment

```python
data.head()
```

```python
sales_profit_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()

color_palette = colors.qualitative.Pastel

fig = go.Figure()
fig.add_trace(go.Bar(x=sales_profit_segment['Segment'],
                    y=sales_profit_segment['Sales'],
                    name = 'Sales',
                    marker_color=color_palette[0]))

fig.add_trace(go.Bar(x=sales_profit_segment['Segment'],
                    y=sales_profit_segment['Profit'],
                    name='Profit',
                    marker_color=color_palette[1]))

fig.update_layout(title='Sales and Profit by Customer Segment',
                 xaxis_title='Customer Segment',yaxis_title='Amount')
fig.show()
```

# sales to profit raito

```python
sales_profit_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sales_profit_segment = ['Sales_to_Ratio'] = sales_profit_segment['Sales'] / sales_profit_segment['Profit']
print(sales_profit_segment[['Segment','Sales_to_Ratio']])
```
